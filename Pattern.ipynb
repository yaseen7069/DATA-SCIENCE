pyramid

```python
rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    print("* " * i)
```

right triangle

```python
rows = 5

for i in range(1, rows + 1):
    print("* " * i)
```

Right triangle using number

```python
rows = 5

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

Inverted Triangle

```python
rows = 5

for i in range(rows, 0, -1):
    print("* " * i)
```
